# Detective Conan: Zero’s Tea Time
## Hit API

```python
hit_conan = requests.request("GET", url_conan)
hit_conan.json()
pre_conan = hit_conan.json()
# pre_conan
```

```python
norm_conan = pre_conan['data']
conan = pd.json_normalize(norm_conan)
conan
```

```python
titles=conan.loc[:,['attributes.title.en']]
desc=conan.loc[:,['attributes.description.en']]
state=conan.loc[:,['attributes.state']]
language=conan.loc[:,['attributes.originalLanguage']]
publicdemographic=conan.loc[:,['attributes.publicationDemographic']]
lastChapter=conan.loc[:,['attributes.lastChapter']]
lastVolume=conan.loc[:,['attributes.lastVolume']]
status=conan.loc[:,['attributes.status']]
rating=conan.loc[:,['attributes.contentRating']]
version=conan.loc[:,['attributes.version']]
```

```python
df_conan = pd.concat([titles,desc,state,language,publicdemographic,lastChapter,lastVolume,status,rating,version],axis=1, ignore_index=False)
df_conan
```

```python
df_conan = df_conan.rename(columns = {'attributes.title.en':'Title','attributes.description.en':'Description','attributes.state':'Statement','attributes.originalLanguage':'Original Language','attributes.publicationDemographic':'Publication Demographic',
                                     'attributes.lastChapter':'Last Chapter','attributes.lastVolume':'Last Volume','attributes.status':'Status','attributes.contentRating':'Rating','attributes.version':'Version'})
df_conan
```

## 1. Detective Conan: Zero’s Tea Time

```python
import requests
import json
endpoint_get1 = 'https://api.jikan.moe/v4/manga?q=Detective Conan: Zero’s Tea Time'
r1 = requests.get(endpoint_get1)
r1.json()

df1 = r1.json()['data']
data1 = pd.DataFrame(df1)
data1

def image_pull(text):
    jpg = text['jpg']
    image = jpg['image_url']
    return image

data1['images'] = data1['images'].apply(lambda x: image_pull(x))

# explode all columns with lists of dicts
df1 = data1.explode("titles").reset_index().drop("index",1).explode("authors").reset_index().drop("index",1)\
        .explode("serializations").reset_index().drop("index",1).explode("genres").reset_index().drop("index",1)\
        .explode("themes").reset_index().drop("index",1).explode("demographics").reset_index().drop("index",1)\
        .explode("title_synonyms").reset_index().drop("index",1)

# list of columns with dicts
cols_to_normalize1 = ['titles', 'authors','serializations','genres','themes','demographics']

# if there are keys, which will become column names, overlap with excising column names
# add the current column name as a prefix
normalized1 = list()
for col in cols_to_normalize1:
    d = pd.json_normalize(df1[col].dropna())
    d.columns = [f'{col}_{v}' for v in d.columns]
    normalized1.append(d.copy())

# combine df with the normalized columns
df1 = pd.concat([df1] + normalized1, axis=1).drop(columns=cols_to_normalize1)

data1 = pd.DataFrame(df1.iloc[13,:])
data1 = np.transpose(data1)
data1
```
